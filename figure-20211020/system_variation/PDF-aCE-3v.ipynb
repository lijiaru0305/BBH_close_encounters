```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
from glob import glob


#import warnings
#warnings.filterwarnings('ignore')
```

```python
## 
##  Read the data for canonical runs
##

RH = (2e-5/3.)**(1./3.)

can_CE0 = np.load('../data/can_CE0-2000.npy')
can_CE1 = np.load('../data/can_mu5x4.npy')
can_CE2 = np.load('../data/can_mu6.npy')

##
##  Plot the time evolution of number
##
plt.rcParams.update({'font.size':22})

fig,axs = plt.subplots(1,1, sharex=True, figsize=(12,12))
fig.subplots_adjust(hspace=0.0)

axs.set_xlabel(r'$R_{\rm H}/a_{\rm rel}$')
axs.set_ylabel(r'PDF')

# parameters of the plot

bin_use = np.linspace(-2.5,2.5,202)
aluse = 0.1
lw = 5

# plot the distribution of RH/a

RH = (2e-5/3.)**(1./3.)
eta = RH/can_CE0[:,3]
Nuse = len(eta)
#bin_use = np.linspace(-3,3,int(2*Nuse**(1./3.)))
axs.hist(eta,bins=bin_use,linewidth=lw,density=True,cumulative=False,histtype='step',
         label=r'fiducial',edgecolor='k',fc=(0,0,0,aluse))

RH = (8e-5/3.)**(1./3.)
eta = RH/can_CE1[:,3]
Nuse = len(eta)
#bin_use = np.linspace(-3,3,int(2*Nuse**(1./3.)))
axs.hist(eta,bins=bin_use,linewidth=lw,density=True,cumulative=False,histtype='step',
         label=r'$(8\times10^{-5},4\times10^{-5})$',edgecolor='r',fc=(1,0,0,aluse))

RH = (2e-6/3.)**(1./3.)
eta = RH/can_CE2[:,3]
Nuse = len(eta)
#bin_use = np.linspace(-3,3,int(np.sqrt(Nuse)))
axs.hist(eta,bins=bin_use,linewidth=lw,density=True,cumulative=False,histtype='step',
         label=r'$(2\times10^{-6},1\times10^{-6})$',edgecolor='b',fc=(0,0,1,aluse))

##  Save the figure
axs.legend(loc=2)

axs.set_xlim([-2.5,2.5])

plt.savefig('./pics/pics-PDF-aCE-3v.pdf',bbox_inches='tight',transparent=True)
plt.savefig('./pics/pics-PDF-aCE-3v.png',bbox_inches='tight',transparent=True)
```
