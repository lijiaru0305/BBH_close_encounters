```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
from glob import glob


#import warnings
#warnings.filterwarnings('ignore')
```

```python
GW_bin = np.load('d:/_work/BBH-data_use/_data/GW/BBH_all.npy')[:,3:6]

RH = (2e-5/3.)**(1./3.)

# parameters of the plot

bin_use = np.linspace(-1.001,1.001,21)
aluse = 0.03
lw = 3

fig,axs = plt.subplots(1,3, sharex=False, figsize=(12,12))
fig.subplots_adjust(hspace=0.0)

## a

#axs[0].set_xlim([-1,1])
#axs.set_ylim([0,1.2])

axs[0].set_xlabel(r'$R_{\rm H}/a_{\rm BBH}$')
axs[0].set_ylabel(r'PDF')

# plot the distribution of cos(ibin)

eta = GW_bin[:,0]*(1-GW_bin[:,1])/RH
Nuse = len(eta)
bin_use = np.linspace(0,100,11)
axs[0].hist(eta,linewidth=lw,density=True,cumulative=False,histtype='step',
         label=r'CE0',edgecolor='k')


eta = GW_bin[:,1]
Nuse = len(eta)
bin_use = np.linspace(0,100,11)
axs[1].hist(eta,linewidth=lw,density=True,cumulative=False,histtype='step',
         label=r'CE0',edgecolor='k')

## inc

axs[2].set_xlim([-1,1])
#axs.set_ylim([0,1.2])

axs[2].set_xlabel(r'$\cos(i_{\rm bin})$')
axs[2].set_ylabel(r'PDF')

# plot the distribution of cos(ibin)

eta = np.cos(GW_bin[:,2])
Nuse = len(eta)
bin_use = np.linspace(-1.001,1.001,11)
axs[2].hist(eta,bins=bin_use,linewidth=lw,density=True,cumulative=False,histtype='step',
         label=r'CE0',edgecolor='k')


##  Save the figure
#axs.legend(loc=9,ncol=2)
#axs.grid()
#axs.set_axisbelow(True)


plt.savefig('./pics/pics-GW-BBH-orbit.pdf',bbox_inches='tight',transparent=True)
plt.savefig('./pics/pics-GW-BBH-orbit.png',bbox_inches='tight',transparent=True)
```
