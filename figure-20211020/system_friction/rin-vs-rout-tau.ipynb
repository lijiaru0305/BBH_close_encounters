```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
from glob import glob


#import warnings
#warnings.filterwarnings('ignore')
```

```python
def scatter_data(df_data,nf_data,snap_list):
    
    plt.rcParams.update({'font.size':22})

    fig,axs = plt.subplots(1,2, sharex=True, sharey=True, figsize=(12,6))
    fig.subplots_adjust(left=None, bottom=None, right=None, top=None, wspace=0.0, hspace=0.)

    alist = np.linspace(0.5,2,100)
    Q = 10**(-5./3.)
    
    K = 2.
    a2list = (1.+K*Q/2.)/(1.-K*Q/2.)*alist
    axs[0].plot(alist,a2list,'k--',lw=2,alpha=0.5)
    axs[1].plot(alist,a2list,'k--',lw=2,alpha=0.5)

    K = 3.
    a2list = (1.+K*Q/2.)/(1.-K*Q/2.)*alist
    axs[0].plot(alist,a2list,'k--',lw=2,alpha=0.5)
    axs[1].plot(alist,a2list,'k--',lw=2,alpha=0.5)
    
    axs[0].set_xlim([0.9,1.03])
    axs[0].set_ylim([0.99,1.12])

    axs[0].set_xlabel(r'$r_{\rm in} = a_{\rm in}(1+e_{\rm in})$  [initial $a_1$]')
    axs[1].set_xlabel(r'$r_{\rm in} = a_{\rm in}(1+e_{\rm in})$  [initial $a_1$]')
    axs[0].set_ylabel(r'$r_{\rm out} = a_{\rm out}(1-e_{\rm out})$  [initial $a_1$]')

    #axs.plot(rin_range,rin_range*1.15,'k',label='stability condition')
    #axs.text(0.91,0.91*1.15+0.003,
    #         r'Petrovich (2015):  lower stability bound $r_{\rm out}/r_{\rm in}\geq1.15$',rotation=37,va='bottom')   
    
    for i in range(len(snap_list)): 

        isnap = snap_list[i]
        
        m1 = df_data[:,isnap,0]
        a1 = df_data[:,isnap,1]
        e1 = df_data[:,isnap,2]
        i1 = df_data[:,isnap,3]

        m2 = df_data[:,isnap,4]
        a2 = df_data[:,isnap,5]
        e2 = df_data[:,isnap,6]
        i2 = df_data[:,isnap,7]

        ain = a1*(a1<a2) + a2*(a1>a2)
        ein = e1*(a1<a2) + e2*(a1>a2)
        aout = a1*(a1>a2) + a2*(a1<a2)
        eout = e1*(a1>a2) + e2*(a1<a2)

        rinapo = ain*(1+ein)
        routperi = aout*(1-eout)

        ratio = routperi/rinapo

        
        axs[1].scatter(rinapo,routperi,alpha=0.3,s=20,label=r'$t=%5d P_1$' %(isnap*4e3))

        
        
        m1 = nf_data[:,isnap,0]
        a1 = nf_data[:,isnap,1]
        e1 = nf_data[:,isnap,2]
        i1 = nf_data[:,isnap,3]

        m2 = nf_data[:,isnap,4]
        a2 = nf_data[:,isnap,5]
        e2 = nf_data[:,isnap,6]
        i2 = nf_data[:,isnap,7]
        
        ain = a1*(a1<a2) + a2*(a1>a2)
        ein = e1*(a1<a2) + e2*(a1>a2)
        aout = a1*(a1>a2) + a2*(a1<a2)
        eout = e1*(a1>a2) + e2*(a1<a2)

        rinapo = ain*(1+ein)
        routperi = aout*(1-eout)

        ratio = routperi/rinapo

        
        axs[0].scatter(rinapo,routperi,alpha=0.3,s=20,label=r'$t=%5d P_1$' %(isnap*4e3))

        
    axs[1].text(0.03, 0.85,r'with drag', ha='left', va='center', transform=axs[1].transAxes)
    axs[0].text(0.97, 0.85,r'no drag', ha='right', va='center', transform=axs[0].transAxes)
    
    axs[0].legend()
    
    plt.savefig('./pics/pics-scatter-rin-vs-rout-tau.pdf',bbox_inches='tight',transparent=True)
    plt.savefig('./pics/pics-scatter-rin-vs-rout-tau.png',bbox_inches='tight',transparent=True)
```
